```python
#2020
```

```python
table_bcn_2020=df_bcn_specie_2020.groupby([pd.Grouper(key="Date_1",freq='M'), "Specie"]).mean()

table_bcn_2020.reset_index(inplace=True)
table_bcn_2020= pd.pivot (table_bcn_2020, values=["max"], columns=["Specie"],index=["Date_1"])

table_bcn_2020
```

```python
table_mad_2020=df_mad_specie_2020.groupby([pd.Grouper(key="Date_1",freq='M'), "Specie"]).mean()

table_mad_2020.reset_index(inplace=True)
table_mad_2020= pd.pivot (table_mad_2020, values=["max"], columns=["Specie"],index=["Date_1"])

table_mad_2020
```

```python
table_bcn_2020.to_csv(path_or_buf= "../data/bcn_specie_2020.csv")
table_mad_2020.to_csv(path_or_buf= "../data/mad_specie_2020.csv")
```
